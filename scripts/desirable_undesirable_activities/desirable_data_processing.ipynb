```python
import pandas as pd
```

```python
df_desirable_rough = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough.csv")
df_desirable_rough_cobb = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough_cobb.csv")
df_desirable_rough_dekalb = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough_dekalb.csv")
df_desirable_rough_fulton = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough_fulton.csv")
df_desirable_rough_gwinnett = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough_gwinnett.csv")
df_desirable_rough_misc = pd.read_csv("../../data/preprocessed/scoring_indicators/ga_desirable_rough_misc.csv")
```

```python
df_combined = pd.concat([df_desirable_rough, df_desirable_rough_cobb, df_desirable_rough_dekalb, 
                         df_desirable_rough_fulton, df_desirable_rough_gwinnett, df_desirable_rough_misc], axis=0)
df_combined = df_combined.reset_index(drop=True)
```

```python
df_final = df_combined.drop_duplicates()
df_final = df_final.reset_index(drop=True)
```
